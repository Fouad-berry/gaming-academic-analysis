# 01 — Exploratory Data Analysis
Shape, distributions, nulls, correlations.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid', palette='muted')
pd.set_option('display.max_columns', None)

df = pd.read_csv('../data/raw/gaming_academic_performance.csv')
print(f'Shape: {df.shape}')
df.head()
```

```python
# Nulls & types
print(df.isnull().sum())
print(df.dtypes)
```

```python
df.describe().T
```

```python
# Categorical distributions
cats = ['gender', 'gaming_genre', 'stress_level']
fig, axes = plt.subplots(1, 3, figsize=(14, 4))
for ax, col in zip(axes, cats):
    df[col].value_counts().plot(kind='bar', ax=ax, color='steelblue')
    ax.set_title(col)
    ax.tick_params(axis='x', rotation=20)
plt.tight_layout()
plt.show()
```

```python
# Correlation heatmap
num_cols = df.select_dtypes('number').columns
plt.figure(figsize=(12, 9))
sns.heatmap(df[num_cols].corr(), annot=True, fmt='.2f', cmap='coolwarm', center=0)
plt.title('Correlation Matrix')
plt.tight_layout()
plt.show()
```

```python
# Grades distribution by genre
fig, axes = plt.subplots(1, 3, figsize=(15, 4), sharey=True)
for ax, genre in zip(axes, df['gaming_genre'].unique()):
    df[df['gaming_genre']==genre]['grades'].hist(bins=30, ax=ax, color='tomato')
    ax.set_title(genre)
    ax.set_xlabel('Grades')
plt.suptitle('Grade Distribution by Gaming Genre')
plt.tight_layout()
plt.show()
```
